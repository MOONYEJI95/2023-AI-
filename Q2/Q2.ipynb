```python
# 클리핑 오류 : 클리핑 에러(Clipping Error)는 주로 음성 또는 오디오 처리와 관련된 개념으로, 소리나 음성 신호를 처리할 때 신호가 최대 허용치를 초과하거나 최소 허용치 미만으로 값을 가질 때 발생하는 오류
```

```python
import librosa
import os
import json
import glob
import sys

def check_audio_files(file_list):
    error_list = []
    for file_name in file_list:
        file_info = {'file_name': file_name, 'error_type': []}

        # 파일 존재 여부 확인
        if not os.path.exists(file_name):
            file_info['error_type'].append('파일이 존재하지 않음')
            error_list.append(file_name)
            continue

        try:
            y, sr = librosa.load(file_name, sr=None)

            # 헤더 유무 확인
            if y.size == 0:
                file_info['error_type'].append('헤더만 존재함')
                error_list.append(file_name)

            # 데이터 유무 및 데이터 값 확인
            else:
                if max(abs(y)) == 0:
                    file_info['error_type'].append('데이터 값이 없음')
                    error_list.append(file_name)

        except Exception as e:
            file_info['error_type'].append('클리핑 에러')
            error_list.append(file_name)

    return error_list


"""
if __name__ == "__main__":
    if len(sys.argv) != 3:
        print("사용법: python3 Q2.py <입력 파일 경로> <출력 파일 경로>")
        sys.exit(1)

    input_file = sys.argv[1]
    output_file = sys.argv[2]
"""


# 'wavlist.txt' 파일에서 경로를 읽어와서 WAV 파일 경로를 glob으로 가져옴
with open('/content/drive/MyDrive/data/wavlist_test.txt', 'r') as wav_file:
    wav_paths = [line.strip() for line in wav_file]

# 'wavlist.txt' 파일에 나열된 WAV 파일 경로를 glob으로 확장
wav_list = []
for wav_path in wav_paths:
    wav_list.extend(glob.glob(wav_path))

# 가져온 WAV 파일 경로를 출력
print("다음 WAV 파일들이 불러와졌습니다:")
for file_path in wav_list:
    print(file_path)

# WAV 파일 오류 검출
error_list = check_audio_files(wav_list)

# 결과를 JSON 파일에 저장
output_file = "/content/drive/MyDrive/data/Q2.json"
result_dict = {"error_list": error_list}

with open(output_file, "w") as json_file:
    json.dump(result_dict, json_file, indent=2)

print(f'오류가 검출된 파일은 {output_file}에 저장되었습니다.')
```
